We can get the number of rows and columns of our dataset through accessing the .shape **attribute** of our DataFrame variable. 

Important, yet pedantic note: 

We should note the difference between ```methods```, which are functions applied to DataFrames, and ```attributes```, which are embedded _properties_ of dataframes that we are merely accessing. 

While ```methods``` have the syntax ```.methodname()```, attributes have the syntax ```.attributename```. 

This is a really common source of errors and confusion, but you can use this as a general guide: 
 - Attributes are like **nouns**. They are existing properties of a DataFrame, or any other object in python. All you're doing is retrieving them - no real functions needed, so no parentheses. 
 - Methods are like **verbs**. They are performing some action on the data, and as such need parentheses, since they're functions.

Or add a new column:

To access a specific row (or rows), we have two main ways of doing things:

````python
df.iloc[0]
````

````python
df.iloc[[0, 10, 35]]
````

````python
df.iloc[0:5]
````

````python
df.loc[0]
````

While .loc may look the same as .iloc and return the same object currently, that is because the labels for each row are the same as their respective indeces. We can see how .loc would function differently if we go to the Pandas documentation that was referred to earlier.
