```python
import pandas as pd
import plotly.express as px
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/silashayes/node-lite-F21/main/women-stem.csv')
print(df.shape)
df.head()
```

```python
df = df.rename(columns={'Median':'Median_Salary'})
df.head()
```

```python
df.isna().sum()
```

```python
df['Share_Women'] = df['Women'] / df['Total']
df.head()
```

```python
fig = px.scatter(df, x='Share_Women', y='Median_Salary')
fig.show()
```

```python
fig1 = px.scatter(df, x='Share_Women', y='Median_Salary', size='Total', color='Major_category', hover_data=['Major'])
fig1.show()
```
