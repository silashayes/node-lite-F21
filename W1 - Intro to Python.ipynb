#### Data Type Functions

These functions tell you the type of a variable or cast (change) a variable into another type.

```
#tells the type of the variable
type(string1)
```

```
string2 = "3"

#casts variables as a specific type
now_str = str(integer1)
now_int = int(string2)

print(now_str)
print(type(now_str))
print(now_int)
print(type(now_int))
```

Important to note: Not all variables can be cast into other variables.
Floats and ints can be cast into each other (with truncation from float to int)
However, a string that looks like a float cannot be cast into an integer, it must first be cast as a float, then can be cast as an integer

```
float2 = 1.9
int2 = 1
str2 = "1.5"

#these functions work
int(float2)
float(str2)

#this function doesn't
int(str2)

#solution
int(float(str2))
```

### Operations (Math)


One thing to note here is how we put str() around the math problems in the print statement. For a print statement to run, everything must be the same variable type, so in this case the integer that results needs to be casted as a string to fit our print statement

```
#addition
print("Addition: 3 + 4 = " + str(3+4))

#multiplication
print("Multiplication: 3 * 4 = " + str(3*4))

#subtraction
print("Subtraction: 3 - 4 = " + str(3-4))

#exponent
print("Exponent: 3^2 = " + str(3**2))

#division
print("Division: 4 / 3 = " + str(4/3))

#int division
print("Int Division: 4 // 3 = " + str(4//3))

#remainder
print("Remainder: 4 % 3 = " + str(4%3))
```

### Comparisons/Conditionals

Comparisons allow us to compare two values and output a boolean

```
print(3>4) #greater than - False
print(3<=4) #less than or equal to - True
print(3==3) #equals - True
print(3!=4) #does not equal - True
```

```
#add logic to comparisons
print(3>4 and 2<3) #and - False
print(3>4 or 2<3) #or - True
print(not 3==3) #not - False
```

Conditionals allow us to build more complex comparisons that can be multileveled, giving us more complex logic

```
x = 4 #change this value and see outcome
if(x>3 and x<6):
  print("Tony")
elif(x==3):
  print("Bennett")
else:
  print("loves you")
```

```
#A different approach
champion = True
if(champion):
  print("Tony Bennett loves you")
else:
  print("Tony Bennett still loves you")
```

##Loops

Loops are the foundation of a lot of computer mathematics. Through them we cna run processes for a specific number of iterations and compute in a way that would take humans years.

```
print(range(10))
print(list(range(10)))
```

```
a = 1
a += 1
a
```

```
a -= 1
a
```

For Loops:

```
num = 0
for i in range(10):
  num += 5
  print("loop #: " + str(i) + ", num: " + str(num))
```

While Loops:

```
credits = 0
while credits < 12:
  print("Part-Time Student with " + str(credits) + " credits")
  credits += 3

print("Full-Time Student with " + str(credits) + " credits")
```

## Defining our own functions

```
def multiply_5(num):
  return num * 5
```

```
multiply_5(10)
```

```
def multiply_num(num, multiplier):
  return num * multiplier
```

```
multiply_num(20, 5.5)
```

```
def do_math(num1, num2):
  if num1 > num2:
    addition = num1 + num2
    subtraction = num1 - num2
    multiplication = num1 * num2
    division = num1 / num2
    return [addition, subtraction, multiplication, division]
```

```
do_math(5, 4)
```
