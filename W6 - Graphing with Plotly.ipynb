```python
import pandas as pd
import plotly.express as px
import matplotlib.pyplot as plt
```

## Plot Creation

https://plotly.com/python/plotly-express/

```python
df = pd.read_csv('https://raw.githubusercontent.com/silashayes/node-lite-F21/main/Week%201/women-stem.csv')
df.head(3)
```

```python
df['Pct_Women'] = df['Women'] / df['Total']
df.head(3)
```

Plotly works in a `px.plot_type` format, where the plot_type dictates the chart type. You can then edit the looks of the plot inline within that `plot_type` function or use the `update_layout()` function to edit the plot after the fact.

```python
fig = px.scatter(df, x='Pct_Women', y='Median')
fig.show()
```

```python
fig = px.scatter(df, x='Pct_Women', y='Median', color='Major_category', size='Total', hover_name='Major')
fig.show()
```

```python
swatches = px.colors.qualitative.swatches()
swatches.show()
```

```python
fig = px.scatter(df, x='Pct_Women', y='Median', color='Major_category', size='Total', hover_name='Major',
                 color_discrete_sequence=px.colors.qualitative.Set1)
fig.show()
```

```python
fig.update_layout(title='Headline Title', xaxis_title='Percent Women', yaxis_title='Median Income (USD)')
```
